```python
import codecs
import pandas as pd
db = dbcon.connect_torndb()
companies = {}
results = []
for line in codecs.open('/Users/eastdog/project/tshbao-git/data/nlp/reports/cachs/ai.code', encoding='utf-8'):
    companies[line.split('\t')[0]] = [db.get('select id from company where code=%s;', code).id for code in line.strip().split('\t')[1:]]
for sector, slist in companies.items():
    for cid in slist:
        item = db.get('select id, companyId, investment, round, currency, precise, fundingDate from funding where (fundingType is null or fundingType=8030) and (active is null or active="Y") and companyId=%s and fundingDate>"2015-12-30;";', cid)
        item['sector'] = sector
        results.append(item)
df = pd.DataFrame(results)
df
```

```python
for k,v in df.groupby('sector'):
    print k, len(v.companyId), v.investment.sum()
```
